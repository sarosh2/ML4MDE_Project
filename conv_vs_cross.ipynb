Generate a Random Input of 5 x 20

```python
import numpy as np
import time
```

```python
np.random.seed(47)

# Generate a 5x20 array of random real numbers
inputs = np.random.rand(5, 20)

print(inputs)
print(inputs.shape)
```

Random weights for 5 neurons

```python
weights = np.random.rand(7, 5)

print(weights)
```

```python
start = time.time()
product = np.dot(weights, inputs)
end = time.time()

print(product)
print(product.shape)
print("Time Elapsed", end - start)
```

```python
repeated_matrix = np.repeat(inputs.T, repeats=7, axis=0)
print(repeated_matrix)
print(repeated_matrix.shape)
```

We first reshape this matrix into shape 20 x 7 x 5. Get 20 chunks of 7 x 5.

Transpose swaps the axis to give 7 x 20 x 5. Then reshape combines the last two axes.

```python
reshaped_matrix = repeated_matrix.reshape(20, 7, 5)
print(reshaped_matrix)
print(reshaped_matrix.shape)
```

```python
start = time.time()
result = (reshaped_matrix * weights).sum(axis=2).T
end = time.time()

print(result)
print(result.shape)
print("Time Elapsed", end - start)
```

```python
print(result / product)
```

```python
import numpy as np
import time

def diff_bw_dot_and_cross(num_features, num_neurons, num_samples, seed):
    
    np.random.seed(seed)
    inputs = np.random.rand(num_features, num_samples)
    weights = np.random.rand(num_neurons, num_features)

    start = time.time()
    product = np.dot(weights, inputs)
    end = time.time()

    p_time = end - start

    start = time.time()
    reshaped_inputs = np.repeat(inputs.T, repeats=num_neurons, axis=0).reshape(num_samples, num_neurons, num_features)
    end = time.time()
    
    reshape_time = end - start

    start = time.time()
    e_product = (reshaped_inputs * weights).sum(axis=2).T
    end = time.time()

    ep_time = end - start

    check = product / e_product
    #print(check, check.sum())
    if check.sum() == num_neurons * num_samples:
        return [p_time, reshape_time, ep_time]
    
    else:
        print("Calculated arrays are unqeual")
        return None
```

```python
results = diff_bw_dot_and_cross(5, 7, 20, 47)

print(results)
print("Reshape Time + Calculation: ", results[1] + results[2])
```

```python
import matplotlib.pyplot as plt

def plot_results(results, num_samples):
    x = np.arange(num_samples)

    # Plot each column as a separate line
    plt.figure(figsize=(10, 6))

    plt.plot(x, results[:, 0], label=f'Matrix Multiplication')
    plt.plot(x, results[:, 1], label=f'Reshape Time')
    plt.plot(x, results[:, 2], label=f'Elementwise Multiplication')
    plt.plot(x, results[:, 3], label=f'Reshape + E_Multiplication')

    # Add labels and title
    plt.xlabel('Num_Samples / 10')
    plt.ylabel('Time in Seconds')
    plt.title('Matrix Multiplication vs Elementwise Multiplication')

    # Add a legend to differentiate the columns
    plt.legend()

    # Show the plot
    plt.show()
```

```python
def remove_outliers(data):
    # Create a mask to track which rows are valid across all columns
    mask = np.ones(data.shape[0], dtype=bool)
    
    # Iterate over each column (algorithm results)
    for i in range(data.shape[1]):
        # Extract the column
        column_data = data[:, i]
        
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = np.percentile(column_data, 25)
        Q3 = np.percentile(column_data, 75)
        
        # Interquartile range (IQR)
        IQR = Q3 - Q1
        
        # Calculate the bounds for outliers
        lower_bound = Q1 - 0.2 * IQR
        upper_bound = Q3 + 0.2 * IQR
        
        # Update the mask: set False for rows where outliers are found
        mask &= (column_data >= lower_bound) & (column_data <= upper_bound)
    
    # Return the filtered array by applying the mask
    cleaned_data = data[mask]
    
    return cleaned_data
```

```python
results = []
for i in range(1000):
    result = diff_bw_dot_and_cross(5, 7, 10 * i, 47)
    result.append(result[1] + result[2])
    results.append(result)

results = np.asarray(results)
print(results.shape)
plot_results(results, 1000)
```

```python
cleaned_results = np.asarray(remove_outliers(results))
print(cleaned_results.shape)
plot_results(cleaned_results, cleaned_results.shape[0])
```

```python
import cupy as cp
import time

def diff_bw_dot_and_cross_gpu(num_features, num_neurons, num_samples, seed):
    
    inputs = cp.random.rand(num_features, num_samples)
    weights = cp.random.rand(num_neurons, num_features)

    # Synchronize before timing the dot product operation
    cp.cuda.Stream.null.synchronize()
    start = time.time()
    product = cp.dot(weights, inputs)
    cp.cuda.Stream.null.synchronize()  # Ensure dot product is complete
    end = time.time()

    p_time = end - start

    # Synchronize before timing the reshaping operation
    cp.cuda.Stream.null.synchronize()
    start = time.time()
    reshaped_inputs = cp.repeat(inputs.T, repeats=num_neurons, axis=0).reshape(num_samples, num_neurons, num_features)
    cp.cuda.Stream.null.synchronize()  # Ensure reshaping is complete
    end = time.time()
    
    reshape_time = end - start

    # Synchronize before timing the element-wise product and sum operation
    cp.cuda.Stream.null.synchronize()
    start = time.time()
    e_product = (reshaped_inputs * weights).sum(axis=2).T
    cp.cuda.Stream.null.synchronize()  # Ensure element-wise product and sum is complete
    end = time.time()

    ep_time = end - start

    # Checking if the calculated arrays are equal
    check = product / e_product
    if cp.allclose(check.sum(), num_neurons * num_samples):  # Ensure correct floating-point comparison
        return [p_time, reshape_time, ep_time]
    
    else:
        print("Calculated arrays are unequal")
        return None
```

```python
import matplotlib.pyplot as plt

def plot_results_gpu(results, num_samples):
    x = np.arange(num_samples)

    # Plot each column as a separate line
    plt.figure(figsize=(10, 6))

    plt.plot(x, results[:, 0], label=f'Matrix Multiplication')
    #plt.plot(x, results[:, 1], label=f'Reshape Time')
    plt.plot(x, results[:, 2], label=f'Elementwise Multiplication')
    #plt.plot(x, results[:, 3], label=f'Reshape + E_Multiplication')

    # Add labels and title
    plt.xlabel('Num_Samples / 10')
    plt.ylabel('Time in Seconds')
    plt.title('Matrix Multiplication vs Elementwise Multiplication')

    # Add a legend to differentiate the columns
    plt.legend()

    # Show the plot
    plt.show()
```

```python
results = []
for i in range(1000):
    result = diff_bw_dot_and_cross_gpu(5, 7, 10 * i, 47)
    result.append(result[1] + result[2])
    results.append(result)

results = np.asarray(results)
print(results.shape)
plot_results_gpu(results, 1000)
```

```python
cleaned_results = np.asarray(remove_outliers(results))
print(cleaned_results.shape)
plot_results_gpu(cleaned_results, cleaned_results.shape[0])
```
